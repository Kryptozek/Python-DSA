```python
# 1.Write a program that asks the user for a number of days. The program then prints out the number of seconds in the amount of days given.

#Loop over the exception until the correct input is received
while True:
    try:
        days = int(input("Please input any number of days "))
        break
    #Catch when the input is not an integer
    except ValueError:
        print("That is not a number,try again  ")

#Store the converted days and output it
seconds = days * 86400
print(f"{days} days is {seconds} seconds")
```

```python
# 2.Write a program that asks a user to input the radius then the program calculates the volume of a sphere (the formula for the volume is (4/3)πr3). Use the exponential operator in python to compute (r3)

import math
#Loop over the exception until the correct input is received
while True:
    try:
        radius = float(input("Please input the radius in numbers "))
        break
    #Catch when the input is not a number
    except ValueError:
        print("That is not a number,try again  ")

#Calculate,store  and output the volume
volume = (4/3) * math.pi * radius**3
print(f"The volume of the sphere is {volume} ")
```

```python
#3. Using functions, write a program to compute the area and perimeter of a square. The program should ask the user to enter a number corresponding to the side length of the square and display the area and perimeter of the square.

def Calculate_area(s):
    area = s**2
    return area

def Calculate_perimeter(s):
    perimeter = s * 4
    return perimeter

def main():
    while True:
        try:
            sidelength = float(input("Please enter the side length of the square "))
            break
        except ValueError:
            print("That is not a number,try again")

    print("The area the square is",Calculate_area(sidelength),"and perimeter is",Calculate_perimeter(sidelength))

main()
```

```python
#4. Write a program using functions that determines whether a character input by a user is uppercase or lower case.


import string
def upper(char):
    status = char.isupper()
    return status


def main():
    while True:
        character = input("Please enter a single character")
        if len(character) == 1 and character.isalpha():
            break
        else:
            print("Invalid input! Please enter a single character (A-Z or a-z only).")

    if upper(character) == True:
        print("The character is uppercase")
    else:
        print("The character is lowercase")

main()
```

```python
#5. The following is pseudocode for a program being designed. Write the Python program equivalent of the same. BEGIN END SET x TO 0, y TO 20 REPEAT SUBTRACT 4 FROM y ADD 2/y TO x UNTIL y IS LESS THAN 6 DISPLAY x
x = 0
y = 20

while True:
    y -= 4
    x += 2 / y

    if y < 6:
        break

print(x)
```

```python
#6. Write a Python program that does the following: • Uses a loop for a user to continually input 5 values to populate an array. • Calculates and displays the average of the values input into the array.

arr = []
for i in range(5):
    val = int(input("Please input a value into an array"))
    arr.append(val)

ans = sum(arr)
print(f"The numbers you are {arr} and the sum is {ans}")
```
